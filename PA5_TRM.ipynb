```python
word="Enter a word to translate from English to Dutch (enter 0 to stop)\n: "
translate={'hello':'Hallo',
           'bye':'doei',
           'dad':'pa',
           'mom':'mama',
           'brother':'broer',
           'sister':'zus',
           'car':'auto',
           'truck':'vrachtwagen',
           'van':'bestelwagen',
           'bird':'vogel',
           'dog':'hond',
           'cat':'kat',
           'chicken':'kip',
           'cow':'koe',
           'horse':'paard',
           'phone':'telefoon',
           'computer':'computer',
           'television':'televisie',
           'camera':'camera',
           'picture':'afbeelding'}
active=True
while active:
    message=input(word)
    if message=='0':
        break
    if message in translate:
        print("The Dutch word for", message,"is", translate[message].upper(),".")
    else:
        print("The Dutch word for", message.upper(), "is not in this dictionary.")
```
